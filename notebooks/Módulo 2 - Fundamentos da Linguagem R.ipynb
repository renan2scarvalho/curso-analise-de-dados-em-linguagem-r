### Estruturas de dados

#### Vetores

Armazena um conjunto de valores ordenados chamados de elementos. Todos os elementos de um vetor devem ser do mesmo tipo

```r
cidades <- c("Brasília", "São Paulo", "Rio de Janeiro")
print(cidades)
print(cidades[1]) # no R, os vetores iniciam em 1
# alterando o vetor
cidades[1] <- "Belo Horizonte"
print(cidades[1:3])
# removendo um elemento
cidades <- cidades[-3]
print(cidades)
```

#### Fator

Armazena valores categóricos (nominal ou ordinal)

```r
UF <- factor(c("DF", "SP", "RJ"))
print(UF)
```

```r
grau.instrucao <- factor(c("Fundamental", "Nível Médio", "Superior", "Nível Médio", "Fundamental"),
                  levels = c("Fundamental", "Nível Médio", "Superior"),
                  ordered=TRUE)
print(grau.instrucao)
```

#### Listas

 É um tipo especial de vetor, porém é possível adicionar elementos de tipos diferentes.

```r
pessoa <- list(sexo="M", cidade="São Paulo", idade=20)
print(pessoa)
print("---")
print(pessoa[["sexo"]])
# deletando elemento via NULL
pessoa[["idade"]] <- NULL
print("---")
print(pessoa)
# adicionando elemento
pessoa[["idade"]] <- 30
print("---")
print(pessoa)
# filtrando elementos
print("---")
print(pessoa[c("cidade", "idade")])
```

#### Dataframe

É utilizado para organizar elementos em linhas e colunas, semelhante a uma planilha ou base de dados. Além disso, pode ser formado por listas ou vetores.

```r
temperatura <- c(30.2, 25.4)
```

```r
df <- data.frame(cidades, temperatura)
print(df)
```

```r
# filtrando dataframe, de maneira semelhante ao df.iloc do pandas
print(df[1, 2])
print("---")
print(df[, 1])
print("---")
print(df[1,])
print("---")
print(df["cidades"])
```

```r
print(names(df)) # colunas
print("---")
print(dim(df)) # dimensão 2x2
print("---")
print(str(df)) # tipos de dados
```

#### Matriz

Armazena dados tabulares semelhante ao data frame, porém só aceita um tipo de dado

```r
m <- matrix(seq(1:25), ncol = 5, byrow = TRUE, dimnames = list(c(seq(1:5)), c("A","B","C","D","E")))
print(m)
```

```r
# filtrando dados
print(m[c(1:2),c("B","C")])
```

### Funções

```r
par.impar <- function(num){
  if ((num%%2) == 0){
    return ("Par")
  } else {
    return ("Ímpar")
  }
}

num <- 3
print(par.impar(num))
num <- 2
print(par.impar(num))
```

### apply

```r
x <- seq(1:9)
m1 <- matrix(x, ncol=3)
print(m1)
```

```r
res <- apply(m1, 1, sum) # 1 -> soma linhas, 2 soma colunas
print(res)
res <- apply(m1, 2, sum)
print(res)
```

```r
numeros.p <- c(2, 4, 6, 8, 10, 12)
numeros.i <- c(1, 3, 5, 7, 9, 11)
numeros <- list(numeros.p, numeros.i)
print(numeros)
lapply(numeros, mean) # média das listas segmentada
sapply(numeros, mean) # média em mesma lista
```

### junções

```r
df1 <- data.frame(Produto=c(1,2,3,4,5), Preco=c(10,20,30,40,50))
df2 <- data.frame(Produto=c(3,5), Nome=c("C","E"))
print(head(df1))
print(head(df2))
```

```r
print(left_join(df1, df2, "Produto"))
```

```r
print(right_join(df1, df2, "Produto"))
```

```r
print(inner_join(df1, df2, "Produto"))
```

### Transformação de dados

```r
df <- data.frame(Produto=c("A","B","C"),
                A.2015 = c(10, 12, 20),
                A.2016 = c(20, 25, 35),
                A.2017 = c(15, 20, 30))
head(df)
```

```r
df2 <- gather(df, "Ano", "Quantidade", 2:3)
head(df2)
```

```r
df2 <- gather(df, "Ano", "Quantidade", 2:4)
head(df2)
```

```r
df3 <- separate(df2, Ano, c("A", "Ano"))
df3 <- df3[-2]
df3
```

```r
df3$Mes <- c("01", "02", "03")
df3
```

```r
df4 <- df3 %>% unite(Data, Mes, Ano, sep="/")
df4
```
